```python
import pyautogui 
import time
import random
import os
import keyboard  # Biblioteca para detectar teclas pressionadas

# Função para ler o arquivo de roteiro
def ler_roteiro(caminho_arquivo):
    with open(caminho_arquivo, 'r', encoding='utf-8') as file:
        linhas = file.readlines()
    return linhas

# Função para abrir o navegador Opera
def abrir_opera():
    # Clica na barra de pesquisa do sistema (coordenadas fornecidas)
    pyautogui.click(111, 1064)  # Coordenadas para clicar na barra de pesquisa do sistema
    time.sleep(0.5)
    pyautogui.write("opera")  # Digita "opera"
    pyautogui.press("enter")  # Pressiona Enter para abrir o Opera
    time.sleep(2)  # Espera o Opera abrir

# Função para buscar a palavra no navegador Opera
def buscar_no_opera(palavra):
    # Clica na barra de pesquisa do Opera (coordenadas fornecidas)
    pyautogui.click(746, 186)  # Coordenadas da barra de pesquisa do Opera
    time.sleep(0.5)
    
    # Limpa o campo de busca anterior (Ctrl + A e Backspace)
    pyautogui.hotkey('ctrl', 'a')  # Seleciona todo o texto
    pyautogui.press('backspace')  # Apaga o texto selecionado
    
    # Digita a nova palavra
    pyautogui.write(palavra)  # Escreve a nova palavra diretamente
    pyautogui.press('enter')  # Faz a pesquisa
    time.sleep(1.5)  # Tempo para a pesquisa carregar
    
    # Clica na aba "Imagens" (coordenadas fornecidas)
    pyautogui.click(392, 247)  # Coordenadas para clicar na aba "Imagens"
    time.sleep(1.5)  # Espera as imagens carregarem

# Função para capturar e salvar o screenshot
def tirar_screenshot(contador, pasta_destino):
    screenshot = pyautogui.screenshot()  # Captura da tela inteira
    caminho_imagem = os.path.join(pasta_destino, f'imagem_{contador}.png')  # Define o caminho do arquivo
    screenshot.save(caminho_imagem)  # Salva a imagem no caminho especificado
    print(f'Screenshot {contador} salvo em {caminho_imagem}')

# Função principal que executa o processo de repetição até obter 300 screenshots ou até a tecla ESC ser pressionada
def executar_script(caminho_arquivo, pasta_destino, total_screenshots=300):
    roteiro = ler_roteiro(caminho_arquivo)  # Lê o roteiro do arquivo
    contador = 0

    # Certifica-se de que a pasta de destino existe
    if not os.path.exists(pasta_destino):
        os.makedirs(pasta_destino)

    abrir_opera()  # Abre o navegador Opera

    # Processo de repetição
    while contador < total_screenshots:
        # Verifica se a tecla ESC foi pressionada para interromper
        if keyboard.is_pressed('esc'):
            print("Automação interrompida pelo usuário.")
            break

        for linha in roteiro:
            if keyboard.is_pressed('esc'):  # Verifica se a tecla ESC foi pressionada
                print("Automação interrompida pelo usuário.")
                return  # Sai imediatamente da função se ESC for pressionado

            palavras = linha.split()  # Divide a linha em palavras
            if not palavras:
                continue
            # Escolhe uma palavra aleatória da linha
            palavra = random.choice(palavras)
            buscar_no_opera(palavra)  # Executa a busca no Opera
            tirar_screenshot(contador, pasta_destino)  # Captura o screenshot
            contador += 1
            if contador >= total_screenshots:
                break

        # Reduzindo o intervalo para 0.5 segundos entre as execuções
        time.sleep(0.5)

# Configurações do script
if __name__ == "__main__":
    caminho_arquivo_txt = "roteiro.txt"  # Coloque aqui o caminho para o arquivo .txt com o roteiro
    pasta_destino_screenshots = "pasta_para_salvar_screenshots"  # Caminho da pasta onde os screenshots serão salvos
    total_de_screenshots = 300  # Número de screenshots desejados

    print("Pressione ESC a qualquer momento para interromper a automação.")
    
    # Executa o script
    executar_script(caminho_arquivo_txt, pasta_destino_screenshots, total_de_screenshots)
```

```python
import pyautogui
import time
import random
import os
import keyboard
import pyperclip  # Biblioteca para manipular o clipboard

# Função para ler o arquivo de roteiro
def ler_roteiro(caminho_arquivo):
    with open(caminho_arquivo, 'r', encoding='utf-8') as file:
        linhas = file.readlines()
    return linhas

# Função para abrir o Google Chrome e configurar a primeira busca
def primeira_tentativa(palavra):
    # Clica no Google
    pyautogui.click(509, 1059)
    time.sleep(0.5)
    
    # Clica no perfil
    pyautogui.click(1053, 606)
    time.sleep(0.5)
    
    # Clica na barra de pesquisa
    pyautogui.click(328, 63)
    time.sleep(0.5)

    # Copia a palavra com "jpg" no final para o clipboard e cola
    pyperclip.copy(palavra + " jpg")
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(0.5)
    
    # Pressiona Enter para buscar
    pyautogui.press('enter')
    time.sleep(1.5)
    
    # Clica na aba de imagens
    pyautogui.click(296, 235)
    time.sleep(1.5)
    
    # Dá zoom na página (Ctrl + duas vezes)
    pyautogui.hotkey('ctrl', '+')
    pyautogui.hotkey('ctrl', '+')
    time.sleep(0.5)
    
    # Clica em uma imagem específica
    pyautogui.click(394, 805)
    time.sleep(1.5)
    
    # Clica com o botão direito do mouse
    pyautogui.click(button='right')
    time.sleep(0.5)
    
    # Clica em "Salvar imagem como" (coordenadas fornecidas)
    pyautogui.click(343, 397)
    time.sleep(0.5)
    
    # Clica em salvar
    pyautogui.click(1209, 720)
    time.sleep(1.5)

# Função para busca e salvar imagem nas tentativas seguintes
def segunda_tentativa(palavra):
    # Clica na barra de pesquisa da aba de imagens
    pyautogui.click(232, 188)
    time.sleep(0.5)

    # Copia a palavra do roteiro para o clipboard e cola
    pyperclip.copy(palavra)
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(0.5)
    
    # Pressiona Enter para buscar
    pyautogui.press('enter')
    time.sleep(1.5)
    
    # Dá zoom na página (Ctrl + duas vezes)
    pyautogui.hotkey('ctrl', '+')
    pyautogui.hotkey('ctrl', '+')
    time.sleep(0.5)
    
    # Clica em uma imagem específica
    pyautogui.click(394, 805)
    time.sleep(1.5)
    
    # Clica com o botão direito do mouse
    pyautogui.click(button='right')
    time.sleep(0.5)
    
    # Clica em "Salvar imagem como"
    pyautogui.click(343, 397)
    time.sleep(0.5)
    
    # Clica em salvar
    pyautogui.click(1209, 720)
    time.sleep(1.5)

# Função para capturar e salvar o screenshot (opcional, se precisar)
def tirar_screenshot(contador, pasta_destino):
    screenshot = pyautogui.screenshot()
    caminho_imagem = os.path.join(pasta_destino, f'imagem_{contador}.png')
    screenshot.save(caminho_imagem)
    print(f'Screenshot {contador} salvo em {caminho_imagem}')

# Função principal que executa o processo de repetição até obter 300 screenshots ou até a tecla ESC ser pressionada
def executar_script(caminho_arquivo, pasta_destino, total_screenshots=300):
    roteiro = ler_roteiro(caminho_arquivo)
    contador = 0

    # Certifica-se de que a pasta de destino existe
    if not os.path.exists(pasta_destino):
        os.makedirs(pasta_destino)

    # Processo de repetição
    while contador < total_screenshots:
        # Verifica se a tecla ESC foi pressionada para interromper
        if keyboard.is_pressed('esc'):
            print("Automação interrompida pelo usuário.")
            break

        for linha in roteiro:
            if keyboard.is_pressed('esc'):
                print("Automação interrompida pelo usuário.")
                return

            palavras = linha.split()
            if not palavras:
                continue

            # Escolhe uma palavra aleatória da linha
            palavra = random.choice(palavras)

            if contador == 0:
                primeira_tentativa(palavra)  # Executa a primeira tentativa
            else:
                segunda_tentativa(palavra)  # Demais tentativas começam na aba de imagens
            
            tirar_screenshot(contador, pasta_destino)  # Captura o screenshot (opcional)
            contador += 1
            
            if contador >= total_screenshots:
                break

        # Intervalo de 0.5 segundos entre execuções
        time.sleep(0.5)

# Configurações do script
if __name__ == "__main__":
    caminho_arquivo_txt = "roteiro.txt"  # Caminho para o arquivo de roteiro
    pasta_destino_screenshots = "pasta_para_salvar_screenshots"  # Pasta para salvar os screenshots
    total_de_screenshots = 300  # Número de screenshots desejados

    print("Pressione ESC a qualquer momento para interromper a automação.")
    
    # Executa o script
    executar_script(caminho_arquivo_txt, pasta_destino_screenshots, total_de_screenshots)
```

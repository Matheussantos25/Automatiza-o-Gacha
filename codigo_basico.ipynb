```python
import pyautogui
import random
import time
import keyboard
import threading  # Para execução em paralelo

# Coordenadas dos personagens e expressões
personagens = [(898, 114), (909, 163), (912, 209), (907, 255), (907, 312), (912, 373)]
coordenadas_face = [(669, 179), (754, 182), (825, 184), (680, 249), (766, 240), 
                    (834, 235), (681, 287), (745, 292), (819, 289), (669, 337),
                    (756, 341), (832, 337), (673, 390), (775, 390), (835, 388),
                    (667, 440), (760, 439), (834, 428), (665, 490), (737, 495)]
coord_mudar_pagina = (831, 526)
coord_mouth_presets = (729, 552)
coord_fechar_presets = (883, 175)
coord_view = (168, 587)

# Função para clicar em uma posição
def clicar(x, y, delay=0.1):
    pyautogui.click(x, y)
    time.sleep(delay)

# Selecionar uma face aleatória
def selecionar_face():
    clicar(842, 251)  # Abre a seção de faces
    if random.choice([True, False]):  # 50% de chance de mudar página
        clicar(*coord_mudar_pagina)
    face = random.choice(coordenadas_face)
    clicar(*face)

# Selecionar uma boca aleatória
def selecionar_mouth():
    clicar(*coord_mouth_presets)
    for _ in range(random.randint(0, 3)):  # Limita a mudança de página para melhorar performance
        clicar(*coord_mudar_pagina)
        time.sleep(0.2)  # Reduz o tempo de espera
    mouth = random.choice(coordenadas_face)
    clicar(*mouth)

# Função para capturar screenshot em thread separada
def capturar_screenshot():
    x, y, largura, altura = 114, 135, 894, 494
    screenshot = pyautogui.screenshot(region=(x, y, largura, altura))
    screenshot.save("screenshot_parcial.png")
    print("Screenshot parcial salva!")

# Execução da automação principal
def executar_automacao():
    clicar(1390, 1060)  # Foca no jogo
    for personagem in personagens:
        if keyboard.is_pressed("esc"):  # Verifica se ESC foi pressionado
            print("Programa interrompido pelo usuário.")
            return
        clicar(*personagem)
        selecionar_face()
        selecionar_mouth()
    
    clicar(*coord_fechar_presets)  # Fecha presets
    clicar(*coord_view)  # Limpa filtros

    # Inicia captura de tela em paralelo
    thread_screenshot = threading.Thread(target=capturar_screenshot)
    thread_screenshot.start()

# Executa a automação com tratamento de interrupção
try:
    executar_automacao()
except KeyboardInterrupt:
    print("Programa interrompido manualmente.")
```

```python
import pyautogui
import time
import pyperclip
import os
import random
import keyboard
import threading

# Função para carregar falas de um arquivo de texto
def carregar_falas_txt():
    falas = []
    with open('falas.txt', 'r', encoding='utf-8') as file:
        for linha in file:
            if linha.strip():  # Ignora linhas vazias
                nome, tempo, fala = linha.split(":")
                falas.append({"nome": nome.strip(), "tempo": float(tempo.strip()), "fala": fala.strip()})
    return falas

# Função para calcular coordenadas relativas à janela do jogo
def calcular_coordenadas_relativas(janela, coord_absoluta):
    # Coordenadas da janela do jogo
    x_janela, y_janela = janela.left, janela.top
    # Converte coordenadas absolutas para relativas à janela
    x_rel, y_rel = coord_absoluta[0] + x_janela, coord_absoluta[1] + y_janela
    return x_rel, y_rel

# Função para clicar em uma coordenada relativa à janela
def clicar_relativo(janela, coord_absoluta, delay=0.5):
    x_rel, y_rel = calcular_coordenadas_relativas(janela, coord_absoluta)
    pyautogui.click(x_rel, y_rel)
    time.sleep(delay)

# Função para copiar e colar o texto no campo de chat
def copiar_colar_texto(texto):
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('enter')

# Função para tirar uma screenshot e salvar em uma pasta organizada
def tirar_screenshot(janela, personagem):
    diretorio_screenshots = 'screenshots_personagens'
    if not os.path.exists(diretorio_screenshots):
        os.makedirs(diretorio_screenshots)
    nome_arquivo = f"{diretorio_screenshots}/{personagem}_{int(time.time())}.png"
    
    # Coordenadas relativas ao topo da janela
    x, y = calcular_coordenadas_relativas(janela, (114, 135))
    largura, altura = 894, 494  # Largura e altura fixas da área do jogo

    # Captura a screenshot usando as coordenadas e dimensões
    screenshot = pyautogui.screenshot(region=(x, y, largura, altura))
    screenshot.save(nome_arquivo)
    return nome_arquivo

# Função para selecionar uma face aleatória
def selecionar_face(janela, coord_mudar_pagina, coordenadas_face):
    clicar_relativo(janela, (842, 251))  # Abre a seção de faces
    if random.choice([True, False]):
        clicar_relativo(janela, coord_mudar_pagina)
    face = random.choice(coordenadas_face)
    clicar_relativo(janela, face)

# Função para selecionar uma boca aleatória
def selecionar_mouth(janela, coord_mudar_pagina, coord_mouth_presets, coordenadas_face):
    clicar_relativo(janela, coord_mouth_presets)
    for _ in range(random.randint(0, 3)):
        clicar_relativo(janela, coord_mudar_pagina)
        time.sleep(0.2)
    mouth = random.choice(coordenadas_face)
    clicar_relativo(janela, mouth)

# Função principal de automação
def automacao_falas():
    falas_personagens = carregar_falas_txt()  # Carrega falas do arquivo .txt
    janela = pyautogui.getWindowsWithTitle("Nome do App")[0]  # Ajuste para o nome correto da janela

    # Log para salvar informações
    log_arquivo = open('automacao_log.txt', 'a')

    # Coordenadas dos personagens relativas ao topo da janela do jogo
    personagens = {
        "Gojo Satoru": (898, 114),
        "Yuji Itadori": (909, 163),
        "Nobara Kugisaki": (912, 209),
        "Maki Zenin": (907, 255),
        "Toge Inumaki": (907, 312),
        "Megumi Fushiguro": (912, 373)
    }

    # Coordenadas das faces e boca
    coordenadas_face = [(669, 179), (754, 182), (825, 184), (680, 249), (766, 240), (834, 235)]
    coord_mudar_pagina = (831, 526)
    coord_mouth_presets = (729, 552)

    # Primeiro clica na posição fixa para focar no jogo
    pyautogui.click(1390, 1060)

    for entrada in falas_personagens:
        nome_personagem = entrada["nome"]
        fala = entrada["fala"]
        tempo = entrada["tempo"]

        # Espera o tempo antes de o personagem falar
        time.sleep(tempo)

        if nome_personagem in personagens:
            coord_personagem = personagens[nome_personagem]

            # Clica no personagem (coordenada relativa)
            clicar_relativo(janela, coord_personagem)

            # Clica no campo de chat (coordenada relativa)
            campo_chat_coord = (848, 179)
            clicar_relativo(janela, campo_chat_coord)

            # Copia e cola a fala
            copiar_colar_texto(fala)

            # Tira a screenshot
            caminho_screenshot = tirar_screenshot(janela, nome_personagem)
            log_arquivo.write(f"{nome_personagem} disse: {fala}\n")
            log_arquivo.write(f"Screenshot salva em: {caminho_screenshot}\n\n")

            # Seleciona uma face e boca aleatória
            selecionar_face(janela, coord_mudar_pagina, coordenadas_face)
            selecionar_mouth(janela, coord_mudar_pagina, coord_mouth_presets, coordenadas_face)

    # Fecha o arquivo de log
    log_arquivo.close()

# Iniciar automação
try:
    automacao_falas()
except KeyboardInterrupt:
    print("Programa interrompido manualmente.")
```

```python
import pyautogui
import pygetwindow as gw
import time

# Função para mover a janela do aplicativo para uma posição específica
def mover_janela(nome_app, x, y):
    try:
        # Obtém a janela pelo nome
        janela = gw.getWindowsWithTitle(nome_app)[0]
        # Move a janela para as coordenadas desejadas
        janela.moveTo(x, y)
        print(f"Janela '{nome_app}' movida para ({x}, {y})")
        time.sleep(1)  # Aguarde um momento para a janela se mover
    except IndexError:
        print(f"Janela '{nome_app}' não encontrada.")

# Função para calcular coordenadas relativas à janela do jogo
def calcular_coordenadas_relativas(janela, coord_absoluta):
    x_janela, y_janela = janela.left, janela.top
    x_rel, y_rel = coord_absoluta[0] + x_janela, coord_absoluta[1] + y_janela
    return x_rel, y_rel

# Função para clicar em uma coordenada relativa à janela
def clicar_relativo(janela, coord_absoluta, delay=0.5):
    x_rel, y_rel = calcular_coordenadas_relativas(janela, coord_absoluta)
    pyautogui.click(x_rel, y_rel)
    time.sleep(delay)

# Coordenadas fixas dos personagens
personagens = {
    "Gojo Satoru": (898, 114),
    "Yuji Itadori": (909, 163),
    "Nobara Kugisaki": (912, 209),
    "Maki Zenin": (907, 255),
    "Toge Inumaki": (907, 312),
    "Megumi Fushiguro": (912, 373)
}

# Função para clicar em um personagem
def clicar_personagem(nome_personagem):
    janelas = gw.getWindowsWithTitle("Gacha Nebula")  # Use o nome correto da janela

    if not janelas:
        print("Janela do aplicativo não encontrada.")
        return

    janela = janelas[0]  # Seleciona a primeira janela encontrada

    if nome_personagem in personagens:
        coord_personagem = personagens[nome_personagem]
        clicar_relativo(janela, coord_personagem)
    else:
        print("Personagem não encontrado.")

# Testando a função
if __name__ == "__main__":
    # Mova a janela do aplicativo para uma posição desejada
    mover_janela("Gacha Nebula", 800, 150)  # Altere para o nome correto do aplicativo e as coordenadas desejadas

    try:
        # Clique no jogo (posição fixa)
        pyautogui.click(1390, 1060)
        time.sleep(1)  # Aguarde um momento para o jogo abrir

        # Clicando em um personagem como exemplo
        clicar_personagem("Gojo Satoru")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")
```

```python
import pyautogui
import time
import random
from PIL import ImageGrab

# Função para mover a janela do jogo
def mover_janela(jogo, x, y):
    pyautogui.getWindowsWithTitle(jogo)[0].moveTo(x, y)

# Função para capturar prints
def capturar_print(nome_arquivo, coordenadas):
    imagem = ImageGrab.grab(bbox=coordenadas)
    imagem.save(f"./prints/{nome_arquivo}.png")

# Função para escolher uma expressão facial e movimento de boca com base na fala
def selecionar_face_mouth(emocao):
    faces = {
        "feliz": [(1501, 330), (1576, 548), (1420, 277), (1511, 376), (1588, 379)],
        "triste": [(1502, 539), (1415, 596), (1498, 540)],
        "neutro": [(1428, 433), (1418, 478), (1581, 437), (1493, 275), (1506, 273), (1566, 280)],
        "surpreso": [(1427, 276), (1422, 331), (1424, 588), (1423, 545), (1580, 489)]
    }
    
    mouths = {
        "feliz": [(1422, 278), (1577, 283), (1419, 330), (1579, 433), (1584, 485), (1579, 537), (1496, 594)],
        "triste": [(1502, 279), (1500, 328), (1416, 492), (1500, 441), (1569, 641)],
        "neutro": [(1504, 543), (1588, 330)],
        "boca_aberta_feliz": [(1584, 279), (1498, 333), (1414, 388), (1589, 374), (1422, 486), (1588, 486), (1428, 573)],
        "boca_aberta_surpreso": [(1586, 331), (1405, 325), (1583, 424)]
    }
    
    # Escolher aleatoriamente coordenadas da face e mouth com base na emoção
    face = random.choice(faces[emocao])
    mouth = random.choice(mouths[emocao])
    
    return face, mouth

# Função para processar cada personagem
def processar_personagem(personagem, coordenada_personagem, falas):
    # Clicar no personagem
    pyautogui.click(coordenada_personagem)
    time.sleep(1)
    
    for fala, emocao in falas:
        # Clicar no Face
        pyautogui.click(1591, 351)
        time.sleep(1)
        
        # Escolher face e mouth com base na emoção
        face_coord, mouth_coord = selecionar_face_mouth(emocao)
        
        # Clicar na face
        pyautogui.click(face_coord)
        time.sleep(1)
        
        # Clicar em Mouth
        pyautogui.click(1492, 653)
        pyautogui.click(mouth_coord)
        time.sleep(1)
        
        # Se a boca estiver aberta, inserir fala no chat
        if "boca_aberta" in emocao:
            pyautogui.click(1595, 279)  # Clicar no chat
            pyautogui.write(fala)       # Escrever fala
            pyautogui.click(1437, 398)  # Sair do chat
            time.sleep(1)
    
    # Fechar o preset de face e mouth
    pyautogui.click(1573, 217)
    time.sleep(1)

# Função principal para processar todos os personagens
def processar_falas(falas):
    personagens = [
        {"nome": "Itadori", "coordenada": (1659, 210)},
        {"nome": "Nobara", "coordenada": (1659, 259)},
        {"nome": "Megumi", "coordenada": (1661, 306)},
        {"nome": "Maki", "coordenada": (1660, 355)},
        {"nome": "Toge", "coordenada": (1657, 403)},
        {"nome": "Gojo", "coordenada": (1657, 450)}
    ]
    
    # Processar cada personagem
    for personagem in personagens:
        processar_personagem(personagem["nome"], personagem["coordenada"], falas)
    
    # Clicar no botão de "view"
    pyautogui.click(859, 631)
    time.sleep(1)

    # Tirar o print
    capturar_print(f"print_{int(time.time())}", (809, 676, 1703, 675))

# Ler falas do arquivo txt e limpar emoções
def ler_falas(arquivo):
    falas = []
    with open(arquivo, "r") as f:
        for linha in f:
            # Separar fala e emoção (esperando algo no formato: "frase (emoção)")
            fala, emocao = linha.rsplit("(", 1)
            emocao = emocao.replace(")", "").replace("\"", "").strip()  # Remover aspas e espaços extras
            falas.append((fala.strip(), emocao))
    return falas

# Mover janela do jogo
mover_janela("Gacha Nebula", 800, 150)
time.sleep(2)

# Clicar na posição inicial (1390, 1060) para ativar o jogo
pyautogui.click(1390, 1060)
time.sleep(2)

# Ler as falas do arquivo
falas = ler_falas("falas.txt")

# Processar as falas para cada personagem
processar_falas(falas)


```

```python
import pyautogui 
import time
import random
import os
from PIL import ImageGrab

# Função para mover a janela do jogo
def mover_janela(jogo, x, y):
    pyautogui.getWindowsWithTitle(jogo)[0].moveTo(x, y)

# Função para capturar prints
def capturar_print(nome_arquivo, coordenadas):
    # Verificar se o diretório "./prints/" existe, caso contrário, criá-lo
    if not os.path.exists("./prints"):
        os.makedirs("./prints")
    
    # Capturar a imagem e salvar no diretório
    imagem = ImageGrab.grab(bbox=coordenadas)
    imagem.save(f"./prints/{nome_arquivo}.png")

# Função para escolher uma expressão facial e movimento de boca com base na emoção
def selecionar_face_mouth(emocao):
    faces = {
        "feliz": [(1501, 330), (1576, 548), (1420, 277), (1511, 376), (1588, 379)],
        "triste": [(1502, 539), (1415, 596), (1498, 540)],
        "neutro": [(1428, 433), (1418, 478), (1581, 437), (1493, 275), (1506, 273), (1566, 280)],
        "surpreso": [(1427, 276), (1422, 331), (1424, 588), (1423, 545), (1580, 489)]
    }
    
    mouths = {
        "feliz": [(1422, 278), (1577, 283), (1419, 330), (1579, 433), (1584, 485), (1579, 537), (1496, 594)],
        "triste": [(1502, 279), (1500, 328), (1416, 492), (1500, 441), (1569, 641)],
        "neutro": [(1504, 543), (1588, 330)],
        "boca_aberta_feliz": [(1584, 279), (1498, 333), (1414, 388), (1589, 374), (1422, 486), (1588, 486), (1428, 573)],
        "boca_aberta_surpreso": [(1586, 331), (1405, 325), (1583, 424)]
    }
    
    # Escolher aleatoriamente coordenadas da face e mouth com base na emoção
    face_coord = random.choice(faces[emocao])
    
    # Se a emoção indicar boca aberta, usar a lista correspondente
    if "boca_aberta" in emocao:
        mouth_coord = random.choice(mouths[emocao])
    else:
        mouth_coord = random.choice(mouths[emocao])
    
    return face_coord, mouth_coord

# Função para processar cada personagem
def processar_personagem(personagem, coordenada_personagem, fala_emocao):
    # Clicar no personagem
    pyautogui.click(coordenada_personagem)
    time.sleep(1)
    
    fala, emocao = fala_emocao
    
    # Clicar no Face
    pyautogui.click(1591, 351)
    time.sleep(1)
    
    # Escolher face e mouth com base na emoção
    face_coord, mouth_coord = selecionar_face_mouth(emocao)
    
    # Clicar na face
    pyautogui.click(face_coord)
    time.sleep(1)
    
    # Clicar em Mouth
    pyautogui.click(1492, 653)
    time.sleep(0.5)
    pyautogui.click(mouth_coord)
    time.sleep(1)
    
    # Se a boca estiver aberta, inserir fala no chat
    if "boca_aberta" in emocao:
        pyautogui.click(1595, 279)  # Clicar no chat
        pyautogui.write(fala)       # Escrever fala
        pyautogui.click(1437, 398)  # Sair do chat
        time.sleep(1)
    
    # Fechar o preset de face e mouth
    pyautogui.click(1573, 217)
    time.sleep(1)

# Função principal para processar todos os personagens
def processar_falas(falas_emocoes):
    personagens = [
        {"nome": "Itadori", "coordenada": (1659, 210)},
        {"nome": "Nobara", "coordenada": (1659, 259)},
        {"nome": "Megumi", "coordenada": (1661, 306)},
        {"nome": "Maki", "coordenada": (1660, 355)},
        {"nome": "Toge", "coordenada": (1657, 403)},
        {"nome": "Gojo", "coordenada": (1657, 450)}
    ]
    
    # Itera pelas falas, usando o índice para associar a cada personagem
    for i, fala_emocao in enumerate(falas_emocoes):
        personagem = personagens[i % len(personagens)]  # Usa o resto da divisão para garantir que as falas sejam atribuídas aos personagens em ciclo
        processar_personagem(personagem["nome"], personagem["coordenada"], fala_emocao)
        
        # Clicar no botão de "view"
        pyautogui.click(859, 631)
        time.sleep(1)

        # Tirar o print (coordenadas ajustadas)
        capturar_print(f"print_{personagem['nome']}_{int(time.time())}", (809, 675, 1703, 676))
        time.sleep(1)  # Aguarda um segundo antes de prosseguir para a próxima iteração
    
# Ler falas do arquivo txt e limpar emoções
def ler_falas(arquivo):
    falas_emocoes = []
    with open(arquivo, "r") as f:
        for linha in f:
            # Separar fala e emoção (esperando algo no formato: "frase (emoção)")
            fala, emocao = linha.rsplit("(", 1)
            emocao = emocao.replace(")", "").replace("\"", "").strip()  # Remover aspas e espaços extras
            falas_emocoes.append((fala.strip(), emocao))
    return falas_emocoes

# Mover janela do jogo
mover_janela("Gacha Nebula", 800, 150)
time.sleep(2)

# Clicar na posição inicial (1390, 1060) para ativar o jogo
pyautogui.click(1390, 1060)
time.sleep(2)

# Ler as falas do arquivo
falas_emocoes = ler_falas("falas.txt")

# Processar as falas para cada personagem
processar_falas(falas_emocoes)

```

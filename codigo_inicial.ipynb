```python
import pygetwindow as gw
import pyautogui
import time

# Espera um pouco para que o usuário possa preparar a janela
print("Prepare a janela que deseja mover. O script irá iniciar em 5 segundos...")
time.sleep(5)

# Obtém a janela ativa
active_window = gw.getActiveWindow()

if active_window is not None:
    print(f"Movendo a janela: {active_window.title}")
    
    # Obtém a posição atual da janela
    x, y = active_window.topleft
    
    # Define a nova posição (por exemplo, 100 pixels para a direita e 100 pixels para baixo)
    new_x = x + 100
    new_y = y + 100
    
    # Move o mouse para a posição da janela e segura o botão
    pyautogui.moveTo(x + 10, y + 10)  # Move para um ponto dentro da janela
    pyautogui.mouseDown()  # Segura o botão do mouse
    
    # Arrasta a janela para a nova posição
    pyautogui.moveTo(new_x, new_y, duration=0.5)  # Duração do movimento
    
    # Libera o botão do mouse
    pyautogui.mouseUp()
else:
    print("Nenhuma janela ativa encontrada.")
```

```python
import pyautogui
import time
import win32gui

coordinates = [
    (1553, 355),
    (1639, 359),
    (1483, 391),
    (1643, 293),
    (913, 720)
]

def screenshot_window(window_title=None):
    """Tira um screenshot da janela especificada."""
    if window_title:
        hwnd = win32gui.FindWindow(None, window_title)
        if hwnd:
            win32gui.SetForegroundWindow(hwnd)
            x, y, x1, y1 = win32gui.GetClientRect(hwnd)
            x, y = win32gui.ClientToScreen(hwnd, (x, y))
            x1, y1 = win32gui.ClientToScreen(hwnd, (x1 - x, y1 - y))
            im = pyautogui.screenshot(region=(x, y, x1, y1))
            return im
        else:
            print(f"Janela '{window_title}' não encontrada.")
            return None
    else:
        return pyautogui.screenshot()

# Substitua "Nome da Janela do Jogo" pelo título correto
window_title = "Gacha Nebula"

while True:
    for x, y in coordinates:
        pyautogui.click(x, y)
        time.sleep(3)

    myScreenshot = screenshot_window(window_title)
    if myScreenshot:
        myScreenshot.save(r'screenshot.png')
        print("Print salvo como screenshot.png")

    time.sleep(3)
```
